# Asiakastietojen muokkaaminen

```python
import pymysql
import mysql.connector as database
import pandas as pd
import numpy as np

db_name = "tietokantaprojekti"
db_host = "localhost"
db_username = "root"
db_passwd = ""
try: conn=pymysql.connect(host=db_host,
                         port=int(3306),
                         user="root",
                         passwd=db_passwd,
                         db=db_name)
except Exception as e:
    print(e)
def updateAsiakas():
    try:
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM asiakas")
        rows = cursor.fetchall()
        column_names = [desc[0] for desc in cursor.description]
        df = pd.DataFrame(rows, columns=column_names)
        print(df)
        
        customer_id = input("Anna päivitettävän asiakkaan asiakasID: ")
        valinta = input("Minkä tiedon haluat päivittää? (etunimi,sukunimi,osoite,email)")
        
        if valinta == "etunimi":
            try:
                v1 = input("Syötä korvaava etunimi: ")
                cursor.execute(f"UPDATE asiakas SET etunimi = %s WHERE asiakasID = {customer_id}", v1)
                conn.commit()
                conn.close()
            except:
                print("Syöttämäsi asiakasID on virheellinen")
            print("Tieto päivitetty tietokantaan")
                
        elif valinta == "sukunimi":
            try:
                v2 = input("Syötä korvaava sukunimi: ")
                cursor.execute(f"UPDATE asiakas SET sukunimi = %s WHERE asiakasID = {customer_id}", v2)
                conn.commit()
                conn.close()
            except:
                print("Syöttämäsi asiakasID on virheellinen")
            print("Tieto päivitetty tietokantaan")
                
        elif valinta == "osoite":
            try:
                v3 = input("Syötä korvaava osoite: ")
                cursor.execute(f"UPDATE asiakas SET osoite = %s WHERE asiakasID = {customer_id}", v3)
                conn.commit()
                conn.close()
            except:
                print("Syöttämäsi asiakasID on virheellinen")
            print("Tieto päivitetty tietokantaan")
                
        elif valinta == "email":
            try:
                v4 = input("Syötä korvaava email")
                cursor.execute(f"UPDATE asiakas SET email = %s WHERE asiakasID = {customer_id}", v4)
                conn.commit()
                conn.close()
            except:
                print("Syöttämäsi asiakasID on virheellinen")
            print("Tieto päivitetty tietokantaan")
        else:
            print("Ei mennyt ihan putkeen, kokeile uudelleen")
            print("")
            print("Sinun tulee syöttää joku näistä valinnoista:")
            print("(etunimi,sukunimi,osoite,email)")
            
            
        
    
    except database.Error as e:
        print(f"Error reading the database: {e}")
        
        
ans = input("haluatko päivittää asiakkaan tiedot? (K/E) (Kyllä/Ei)")
kyl = ["Kyllä", "K", "Joo", "Yes", "Jep", "KYLLÄ", "joo", "k","kyllä","k'","K'"]
ei = ["En", "Ei", "Älä", "älä", "EI", "ei","en"]
if ans in kyl:
    updateAsiakas()
else:
    print("Kokeile uudelleen")
```

# Asiakkaan poistaminen

```python
import pymysql
import mysql.connector as database
import pandas as pd
import numpy as np

db_name = "tietokantaprojekti"
db_host = "localhost"
db_username = "root"
db_passwd = ""

try: conn=pymysql.connect(host=db_host,
                         port=int(3306),
                         user="root",
                         passwd=db_passwd,
                         db=db_name)
except Exception as e:
    print(e)
    
def delAsiakas():
    try:
 
        cursor = conn.cursor()
    
        cursor.execute("SELECT * FROM asiakas")
        rows = cursor.fetchall()
        column_names = [desc[0] for desc in cursor.description]
        df = pd.DataFrame(rows, columns=column_names)
        print(df)
        
        customer_id = input("Anna asiakkaan asiakasID minkä haluat poistaa: ")
        cursor.execute(f"DELETE FROM asiakas WHERE asiakasID = {customer_id}")
            
        conn.commit()
        conn.close()
    
    except database.Error as e:
        print(f"Error reading the database: {e}")

ans = input("haluatko poistaa asiakkaan? (K/E) (Kyllä/Ei)")
kyl = ["Kyllä", "K", "Joo", "Yes", "Jep", "KYLLÄ", "joo", "k"]
ei = ["En", "Ei", "Älä", "älä", "EI", "ei","en"]
if ans in kyl:
    delAsiakas()
    print("Asiakas poistettu onnistuneesti.")
else:
    print("Selvä homma, ensi kertaan!")
```
